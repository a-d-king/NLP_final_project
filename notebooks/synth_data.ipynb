````python
import json
import pandas as pd

def load_and_clean_sentences(file_path):
    """
    Load and clean sentences from a JSON file with a simple list format.
    
    Args:
    file_path (str): Path to the JSON file.
    
    Returns:
    pd.DataFrame: DataFrame containing cleaned sentences.
    """
    with open(file_path, 'r') as file:
        data = json.load(file)

    all_sentences = []
    for sentences in data.values():
        for sentence in sentences:
            if sentence not in ["```json", "[", "]", "```"]:
                # Remove escaping of double quotes and strip unwanted characters
                cleaned_sentence = sentence.strip().replace('\\"', '"')
                all_sentences.append(cleaned_sentence)

    df_sentences = pd.DataFrame(all_sentences, columns=['notes'])

    # Improved regex to handle quotation marks at both ends, including optional punctuation
    df_sentences['notes'] = df_sentences['notes'].str.replace(r'^"(.*)"[,.;:"]*$', r'\1', regex=True)

    return df_sentences


# Example usage for both types of files
path_to_json1 = '/Users/tarunvallabhaneni/nlp-final-project/generated_sentences.json'
path_to_json2 = '/Users/tarunvallabhaneni/nlp-final-project/generated_subtopic_sentences.json'

df_sentences1 = load_and_clean_sentences(path_to_json1)
df_sentences2 = load_and_clean_sentences(path_to_json2)

# Display the first few rows of the DataFrame to confirm
print("First File - Sentences DataFrame:")
print(df_sentences1.head())
print("\nSecond File - Sentences DataFrame:")
print(df_sentences2.head())
````
